```python
import pandas as pd
import numpy as np
import random
import string

# index = ['aaaaa' for i in range(10)]

'''
pd.DataFrame(
    [{i: np.random.randint(100 + 1) for i in columns} for j in index],
    index=index
)
'''

def generate():
    return ''.join(random.choice(string.ascii_lowercase+string.ascii_uppercase) for i in range(5))

index = [generate() for i in range(10)]
columns = ['국어', '영어', '수학', '사회']

pd.DataFrame([[np.random.randint(100 + 1) for i in columns] for j in index], index=index, columns=columns)
```
