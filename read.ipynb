```python
import pandas as pd
import numpy as np
import string
import random

df = pd.DataFrame(np.random.randint(0, 100, size=(10, 4)),
             index=list((''.join(random.choice(string.ascii_letters) for i in range(5))) for i in range(10)),
             columns=['국어', '영어', '수학', '사회']
            )
df
```

```python
df.loc[:, ['국어', '영어']]
```

```python
# 과학 추가. 그에 따라 과학에도 점수 입력
# df['과학'] = [np.random.randint(100 + 1) for i in range(10)]
df.loc[:, '과학'] = pd.Series([np.random.randint(100 + 1) for i in range(10)], index=df.index)
df
```

```python
# 각 학생들의 점수의 총합을 표현하는 컬럼 추가
df['개인 총합'] = df.sum(axis=1)
df
```

```python
ls = df.sum().tolist()
ls
```

```python
df.loc['과목총점'] = ls
df
```
